```python
# 미국 Boston 지역의 집값 data 불러오기
import pandas as pd
house = pd.read_csv('https://raw.githubusercontent.com/rasbt/'
    'python-machine-learning-book-2nd-edition'
    '/master/code/ch10/housing.data.txt', header=None, sep='\s+')
house.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
                 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
house.head()
```

```python
# 종속변수 MEDV(집가격)와 설명변수의 Plot 그리기.
import plotly.express as px
fig = px.scatter_matrix(house,
    dimensions=['LSTAT', 'INDUS', 'NOX', 'RM', 'MEDV'],
    title='Scatter Matrix')
fig.show()
fig = px.scatter_matrix(house,
    dimensions=['ZN', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'MEDV'],
    title='Scatter Matrix')
fig.show()
```

```python
# 종속변수 MEDV(집가격)와 설명변수의 상관계수 행렬 구하기
import numpy as  np
col1=['LSTAT', 'INDUS', 'NOX', 'RM', 'MEDV']
col2=['ZN', 'AGE', 'DIS', 'TAX', 'PTRATIO', 'MEDV']
a=house[col1].values
print(a.shape)
cm1 = np.corrcoef(house[col1].values.T)
cm2 = np.corrcoef(house[col2].values.T)
print(cm1)
print(cm2)
```

```python
# 일부 변수에 대한 log 변환
import numpy as np
house['LLSTAT']=np.log(house['LSTAT'])
house['LINDUS']=np.log(house['INDUS'])
house.head()
```

```python
# 전체 data를 종속변수 y와 특성변수 X의 data로 나누기
y = house['MEDV'].values
house1=house.drop(['LSTAT', 'INDUS', 'MEDV'], axis=1)
X = house1.values
```

```python
# 전체 data를 training data(70%)와 test data(30%)로 나누기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
    test_size=0.3, random_state=1)
```

```python
# regression module 불러오기와 모형추정
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(X_train, y_train)
```

```python
# 추정값 확인
print('Slpoe:', mlr.coef_)
print('Intercept:' , mlr.intercept_)
```

```python
# 예측치 구하기
y_train_pred = mlr.predict(X_train)
y_test_pred = mlr.predict(X_test)
```

```python
# 잔차에 대한 plot 그리기.
import plotly.graph_objects as go
fig=go.Figure()
fig.add_trace(go.Scatter(x=y_train_pred, y=y_train_pred-y_train,
    mode='markers', name='Training data'))
fig.add_trace(go.Scatter(x=y_test_pred, y=y_test_pred-y_test,
    mode='markers', name='Test data'))
fig.update_layout(width=600, height=400,
    title_text='Residual Plots versus predicted values', title_x=0.5)
fig.update_xaxes(title_text='predicted')
fig.update_yaxes(title_text='residuals')
fig.show()
```

```python
# 모형의 MSE
from sklearn.metrics import mean_squared_error
print('MSE train: %.3f, test: %.3f' %
    (mean_squared_error(y_train, y_train_pred),
    mean_squared_error(y_test, y_test_pred)))
```

```python
# 모형의 R^2
from sklearn.metrics import r2_score
print('R^2 train: %.3f, test: %.3f' %
    (r2_score(y_train, y_train_pred),
    r2_score(y_test, y_test_pred)))
```

```python
########### 로버스트 회귀 'RANSAC Regressor'
from sklearn.linear_model import RANSACRegressor
rans = RANSACRegressor(max_trials=100, min_samples=45,
    loss='absolute_error', residual_threshold=5.0, random_state=1)
rans.fit(X_train, y_train)
y_train_pred=rans.predict(X_train)
y_test_pred=rans.predict(X_test)
print(rans.estimator_.coef_)
print(rans.estimator_.intercept_)
```

```python
# 모형의 MSE training set
mean_squared_error(y_train, y_train_pred)
```

```python
# 모형의 MSE test set
mean_squared_error(y_test, y_test_pred)
```

```python
# RANSAC에서 회귀 계수를 추정하는데 사용한 데이터(inner)와
# 특이치(outlier) 출력
inlier_mask=rans.inlier_mask_
print('inner', inlier_mask)
```

```python
outlier_mask=np.logical_not(inlier_mask)
print('outlier', outlier_mask)
```

```python
############# SVR.Regression......
from sklearn.svm import SVR # SVR module import
svl=SVR(kernel = 'linear', C=1.0, epsilon=0.1) # 선형 SVM회귀
svr=SVR(kernel = 'rbf', C=1.0, epsilon=0.1) # 비선형 SVM회귀
svl.fit(X_train, y_train)
svr.fit(X_train, y_train)
```

```python
print(svl.support_)
print(svl.coef_)
print(svl.intercept_)
```

```python
# SVM 회귀 예측치 계산
y_train_predsvl=svl.predict(X_train)
y_train_predsvr=svr.predict(X_train)
y_test_predsvl=svl.predict(X_test)
y_test_predsvr=svr.predict(X_test)
```

```python
# 모형의 MSE
from sklearn.metrics import mean_squared_error
mse_l_trina=mean_squared_error(y_train, y_train_predsvl)
mse_l_test=mean_squared_error(y_test, y_test_predsvl)
mse_n_train=mean_squared_error(y_train, y_train_predsvr)
mse_n_test=mean_squared_error(y_test, y_test_predsvr)
print(mse_l_trina, mse_l_test, mse_n_train, mse_n_test)
```

```python
# 모형의 R^2
from sklearn.metrics import r2_score
R2_l=r2_score(y_train, y_train_predsvl)
R2_n=r2_score(y_train, y_train_predsvr)
print(R2_l, R2_n)
```
