```python
import numpy as np
import plotly.graph_objects as go
np.random.seed(0)
N=100
x1=np.linspace(0,10,N)
y1=5.+x1+np.random.randn(N)
y2=x1+np.random.randn(N)
y3=-5.+x1+np.random.normal(0,2,N)

fig=go.Figure()
fig.add_trace(go.Scatter(x=x1, y=y1, mode='lines', name='line'))
fig.add_trace(go.Scatter(x=x1, y=y2, mode='lines+markers', name='line+marker'))
fig.add_trace(go.Scatter(x=x1, y=y3, mode='markers', name='marker'))

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

np.random.seed(0)
N=100
x=np.linspace(-4,4,N)
y1=1/np.sqrt(2*np.pi*4)*np.exp(-x**2/(2*4))
y2=1/np.sqrt(2*np.pi*1)*np.exp(-(x-0.3)**2/(2*1))
y3=1/np.sqrt(2*np.pi*1)*np.exp(-(x+0.5)**2/(2*1))

fig=go.Figure()
fig.add_trace(go.Scatter(x=x,y=y1,mode='lines', name='line'))
fig.show()
fig=go.Figure()
fig.add_trace(go.Scatter(x=x,y=y2,mode='lines', name='line'))
fig.show()
fig=go.Figure()
fig.add_trace(go.Scatter(x=x,y=y3,mode='lines', name='line'))
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
np.random.seed(0)
N=100

x1=np.random.normal(1,1,N)
x1=np.sort(x1)
x2=np.random.randn(N)
x2=np.sort(x2)
y1=5.+x1-2*x2+np.random.normal(0,4,N)
y2=-4*x1-2*x2**3
y3=5.+x1-3*x2+np.random.normal(0,2,N)

surface_1=go.Scatter3d(x=x1,y=x2,z=y1)
surface_2=go.Scatter3d(x=x1,y=x2,z=y2)
surface_3=go.Scatter3d(x=x1,y=x2,z=y3)
data=[surface_1,surface_2,surface_3]
fig=go.Figure(data=data)
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
np.random.seed(0)
N=100

x1,x2=np.meshgrid(np.linspace(0,2,N),np.linspace(0,2,N))
print(x1.shape)
y1=5.+x1+2*x2
y2=y1.copy()+np.random.normal(0,2,(N,N))

surface_1=go.Surface(x=x1,y=x2,z=y1)
surface_2=go.Scatter3d(x=x1.flatten()[::20],
                       y=x2.flatten()[::20],z=y2.flatten()[::20],
                       opacity=0.3, mode='markers')
data=[surface_1,surface_2]
fig=go.Figure(data=data)
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import numpy as np
fig=make_subplots(rows=1, cols=2)
X=np.linspace(-3,3,20)
Y1=np.sin(X)
Y2=np.cos(X)
Y3=np.tan(X)
fig.add_trace(go.Scatter(x=X,y=Y1),row=1,col=1)
fig.add_trace(go.Scatter(x=X,y=Y2),row=1,col=2)
fig.update_layout(height=400,width=800,
                  title_text='side by side plots',title_x=0.5)
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
fig=make_subplots(rows=2, cols=1)
X=np.linspace(-3,3,20)
X1=np.linspace(-5,5,30)
X2=np.linspace(-10,10,50)
Y1=np.sin(X)
Y2=np.cos(X)
Y3=np.tan(X)
Y4=np.tanh(X)
fig.add_trace(go.Scatter(x=X,y=Y1), row=1,col=1)
fig.add_trace(go.Scatter(x=X,y=Y2), row=1,col=1)
fig.add_trace(go.Scatter(x=X,y=Y3), row=2,col=1)
fig.update_layout(height=500, width=500,
                  title_text='Stack plots', title_x=0.5)
fig.show()
```

```python
fig=make_subplots(rows=2,cols=2,
                  subplot_titles=('sin','tan','cos','tanh'),
                  row_heights=(0.7,0.3), column_width=(0.7,0.3))
fig.add_trace(go.Scatter(x=X,y=Y1), row=1,col=1)
fig.add_trace(go.Scatter(x=X1,y=Y3), row=1,col=2)
fig.add_trace(go.Scatter(x=X1,y=Y2), row=2,col=1)
fig.add_trace(go.Scatter(x=X2,y=Y4), row=2,col=2)
fig.update_layout(height=500, width=800,
                  title_text='multiple curves', title_x=0.5)
fig.show()
```

```python
fig=make_subplots(rows=2,cols=2,
                  subplot_titles=('sin','tan','cos','tanh'),
                  vertical_spacing=0.1, horizontal_spacing=0.3)
fig.add_trace(go.Scatter(x=X,y=Y1), row=1,col=1)
fig.add_trace(go.Scatter(x=X1,y=Y3), row=1,col=2)
fig.add_trace(go.Scatter(x=X1,y=Y2), row=2,col=1)
fig.add_trace(go.Scatter(x=X2,y=Y4), row=2,col=2)
fig.update_xaxes(title_text='x',col=1,row=1)
fig.update_xaxes(title_text='x1',col=2,row=1)
fig.update_xaxes(title_text='x1',col=1,row=2)
fig.update_xaxes(title_text='x2',col=2,row=2)
fig.update_xaxes(title_text='y1',col=1,row=1)
fig.update_xaxes(title_text='y3',col=2,row=1)
fig.update_xaxes(title_text='y2',col=1,row=2)
fig.update_xaxes(title_text='y4',col=2,row=2)
fig.update_layout(height=500, width=800,
                  title_text='customized multiple curves', title_x=0.5)
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
fig=make_subplots(rows=1,cols=2)
fig.add_trace(go.Scatter(x=[1,2,3],y=[4,5,6],
                         mode='markers+text',text=['text A','text B','text C'],
                         textposition='bottom center'), row=1, col=1)
fig.add_trace(go.Scatter(x=[10,20,30],y=[40,50,60],
                         mode='markers+text',text=['text Aa','text Bb','text Cc'],
                         textposition='bottom center'), row=1, col=2)
fig.update_layout(height=600, width=900,
                  title_text='plots with annotations')
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
fig=make_subplots(rows=2, cols=2, 
                  specs=[[{},{}],[{'colspan':2},None]],
                  subplot_titles=('F1','F2','F3'))
fig.add_trace(go.Scatter(x=[1,2,3],y=[4,5,6]), row=1, col=1)
fig.add_trace(go.Scatter(x=[10,20,30],y=[40,50,60]), row=1, col=2)
fig.add_trace(go.Scatter(x=[1,2,3],y=[4,1,6]), row=2, col=1)
fig.update_layout(height=600, width=900,
                  title_text="plots with annotations")
fig.show()
```
